```python
from sklearn.datasets import make_classification
import numpy as np
x,y = make_classification(n_samples=100,n_features=2,n_informative=2,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=10)
print(x.shape,y.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
plt.show()
```

```python
def step(z):
  return 1 if z>0 else 0
```

```python
def perceptron(x,y):
  x = np.insert(x,0,1, axis = 1)
  weights = np.ones(x.shape[1])
  lr = .00001

  for i in range(1000):
    j = np.random.randint(0,100)
    y_hat = step(np.dot(x[j],weights))
    error = y[j] - y_hat
    weights = weights + lr * error * x[j]
    # print("weights",weights[0],weights[1:])
  return weights[0],weights[1:]
```

```python
intercept_,coef_ = perceptron(x,y)
```

```python
print(intercept_,coef_)
```

```python
m = -coef_[0]/coef_[1]
b = intercept_/coef_[1]
```

```python

X_input = np.linspace(-3,3,100)
Y_input = m*X_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(X_input,Y_input,'r')
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
```
